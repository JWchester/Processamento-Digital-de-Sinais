#Biblioteca

```python
import numpy as np
import plotly.graph_objects as go
```

#Funções

```python
# adiciona 0´s a esquerda dos bits
def adiciona_zeros(x):
  for i, j in enumerate(x):
    while (len(x[i]))< np.log2(len(x)):
      x[i] = "0" + x[i]
  return x

# Inverte os bits de cada valor do vetor
def inverter_bits(x):
  return [s[::-1] for s in x]

#converte o valor binario para decimal
def base2_p_base10(x):
  return [int(s,2) for s in x]

#reordena lista após a troca de indices
def reordena(x_indices,x_valores):
  return [x_valores[i] for i in x_indices]

# calcula a transformada rápida de fourier
def FFT(x,N):

  if N == 1:
    return x
  else:
    Xp = FFT(x[N//2:], N//2)
    Xi = FFT(x[:N//2], N//2)
    X = np.zeros(N, dtype=complex)
    W_N = np.exp(-1j * 2 * np.pi/N)

  for k in range(0,N//2):
    u = W_N**k * Xi[k]
    X[k] = Xp[k] + u
    X[k + N//2] = Xp[k] - u

  return X


# Adiciona zeros para completar a matriz com N elementos(aumentando a precisão do gráfico)
def zero_padding(N,x):
  while len(x) < N:
    x.append(0)

  return x


# plot de gráficos interativos
def stem_plot_plotly(x, y, title='Stem Plot', y_label='Amplitude'):
    fig = go.Figure()

    # Linhas verticais
    for xi, yi in zip(x, y):
        fig.add_trace(go.Scatter(
            x=[xi, xi],
            y=[0, yi],
            mode='lines',
            line=dict(color='blue'),
            showlegend=False
        ))

    # Marcadores no topo das hastes
    fig.add_trace(go.Scatter(
        x=x,
        y=y,
        mode='markers',
        marker=dict(color='blue', size=8),
        name='valor'
    ))

    # Linha de base
    fig.add_trace(go.Scatter(
        x=[x[0], x[-1]],
        y=[0, 0],
        mode='lines',
        line=dict(color='red', width=1),
        showlegend=False
    ))

    # atualiza gráfico
    fig.update_layout(
        title=title,
        xaxis_title='(k)',
        yaxis_title=y_label,
        height=400
    )

    fig.show()
```

#IMPLEMENTAÇÃO

```python
#Número de amostras
N = 512


#sinal de entrada(+ ajuste de dimensão caso o vetor seja diferente de N)

x = zero_padding(N,[1,1,1,1,0,0,0,0])

# recebe os indices do sinal e transforma em bits
x_indices_bits = [bin(i)[2:] for i in range(len(x))]

# inverte os bits dos indices
x_new_indices = base2_p_base10(inverter_bits(adiciona_zeros(x_indices_bits)))

# reorganiza  os valores do sinal de acordo com os novos índices
x_reordenados = reordena(x_new_indices,x)


# aplica transformada de fourier
fft = FFT(x_reordenados,N)
```

#GERAÇÃO DE GRAFICOS DE MODULO E FASE

```python
# Plotar módulo interativos
k = np.arange(N)
stem_plot_plotly(k, np.abs(fft), title='Módulo da FFT', y_label='|X[k]|')

# Plotar fase interativos
stem_plot_plotly(k, np.angle(fft), title='Fase da FFT', y_label='∠X[k] (rad)')
```

##Algoritmos Típicos que utilizam recursividade

```python
# sequência de fibonacci

def fib(n):
  if n ==1:
    return 1
  elif n == 0:
    return 0
  else:
    return fib(n-2) + fib(n-1)

print(fib(1))


# fatorial
def fatorial(n):

  if n == 1:
    return 1
  else:
    return n * fatorial(n-1)

print(fatorial(3))
```
